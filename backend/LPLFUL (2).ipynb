```python




from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False, drop='first'),['batting_team','bowling_team','city'])
]
,remainder = 'passthrough')
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

pipe1 = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])

pipe1.fit(X_train,y_train)
y_pred = pipe1.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

pipe2 = Pipeline(steps=[
    ('step1',trf),
    ('step2',RandomForestClassifier())
])

pipe2.fit(X_train,y_train)
y_pred = pipe2.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

pipe3 = Pipeline(steps=[
    ('step1',trf),
    ('step2',DecisionTreeClassifier())
])

pipe3.fit(X_train,y_train)
y_pred = pipe3.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score

def evaluate_regression(model, X_test, y_test):
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)

    print("Regression Report")
    print("Accuracy:", acc)
    print("================================================================")
```

```python
pipes = {
    "LogisticRegression": pipe1,
    "RandomForestClassifier": pipe2,
    "DecisionTreeClassifier": pipe3
}
for name, model in pipes.items():
    print(name)
    model.fit(X_train, y_train)
    #tuning
    evaluate_regression(model, X_test, y_test)
```

```python
def match_progression(x_df,match_id,pipe1):
  match = x_df[x_df['match_id'] == match_id]
  temp_df = match[['batting_team','bowling_team','city','runs_left','balles_left','wickets','total_runs_x','crr','rrr','result']]
  temp_df = temp_df[temp_df['balles_left'] != 0]
  result = pipe1.predict_proba(temp_df)
  temp_df['lose'] = np.round(result.T[0]*100,1)
  temp_df['win'] = np.round(result.T[1]*100,1)
  temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)

  target = temp_df['total_runs_x'].values[0]
  runs = list(temp_df['runs_left'].values)
  new_runs = runs[:]
  runs.insert(0,target)
  temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
  wickets = list(temp_df['wickets'].values)
  wickets_in_over = [wickets[0]] + [wickets[i] - wickets[i - 1] for i in range(1, len(wickets))]
  temp_df['wickets_in_over'] = wickets_in_over


  print("Target-",target)
  temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over', 'lose','win']]
  return temp_df,target
```

```python
temp_df,target = match_progression(delivery_df,1238760,pipe1)
temp_df
```
